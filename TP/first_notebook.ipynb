```python
import pandas as pd
data = pd.read_csv("ecological_footprint_consommation.csv",encoding = 'ISO-8859-1', index_col=0)
data.head()

#Import des autres modules utiles
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


n=data.shape[0]
print('Nombre d\'individus (pays) :', n)

## La première variable contient la région du globe.
## Récupérer la donnée qualitative (la région du globe)
region=data.loc[:,data.columns[0]].to_list()


## Récupérer le tableau des données quantitatives

#considérer les données de la colonne start_col à la dernière colonne
start_col=1 #doit être entre 1  (la colonne 0 est celle de la région) et data.shape[1]

p=data.shape[1]-start_col
print('La dimension des données est : ', p)

X=data.loc[:, data.columns[start_col]:data.columns[start_col+p-1]].to_numpy().reshape([n,p])
print('Taille du tableau des données :', X.shape,'\n')

#Exemples
print('L\'individu \"', data.index[4],'\" a pour valeur de région :',region[4],'\n')
print('L\'individu \"', data.index[4],'\" a pour variables :\n', X[4,:],'\n')

print('Les valeurs de la variable \"',  data.columns[3] , '\" pour l\'ensemble des individus sont:\n', X[:,2])
#décalage d'index ci dessus, car la première colonne a été mise dans région
```

```python
                        #Données de base :
    
n=data.shape[0]            #individus
start_col=1
p=data.shape[1]-start_col  #variables
    
X=data.loc[:, data.columns[start_col]:data.columns[start_col+p-1]].to_numpy().reshape([n,p])
   

                      #calcul du centrage de données

s1=0        
U=np.zeros([n,p])
for i in  range(n):
    for j in  range(p):
            for k in  range(n):
                s1=s1+X[k,j]
            U[i,j] =X[i,j]-(1/n)*s1

                    #calcul de la matrice de covariance

C=(1/n)*np.dot(U.T,U)
#print(C)

                    #calcul des vecteurs et valeurs propres

W0=np.linalg.eig(C)[0]
#print(W0.shape)
#print(W0)
W1=np.linalg.eig(C)[1]
print("")

#print(W0) #ensemble des valeurs propres
print(W1) #ensemble des vecteurs propres


                    #Ordonner les valeurs propres

W0_t=[]
W0_bis=np.linalg.eig(C)[0]
r = 0
for k in range(p):
    m=max(W0_bis[r:])
    #debut calcul indice max[r:]
    j = 0
    i=-1
    while j < p and i == -1:
        if W0[j]==m:
            i=j
        j+=1
        #fin indice max[r:]
    W0_t.append(i)
    W0_bis[r],W0_bis[i]=W0_bis[i],W0_bis[r]
    r += 1
#print(W0_t)


                    #calcul de l'inertie relative

s3=0 #somme des valeurs propres
for i in W0:
    S3=s3+i

I=np.zeros(p)
for i in range(p):
    I[i]=W0[i]/S3
#print(I)

W0_propI = np.zeros(p)
for i in range(p):
    W0_propI[i]=W0[i]/I[0]

#print(W0.shape)
#print(I)

#Décomposition de Cov en vecteur propre/valeurs propres

A = np.diag(W0)
print(A)
Wf= np.dot(np.dot(W1.T,A),W1)
Wf1=np.dot(W1.T,A)
Wf2=np.dot(Wf1,W1)
print(" ")
print(Wf2)

```
